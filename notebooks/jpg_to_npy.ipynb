```python
import os
import numpy as np
from PIL import Image
from tqdm import tqdm
```

```python
jpg_dir = "/home/ilya/Documents/IT-Geo-Paleochannel/data/river_dataset_test_1/masks"
npy_dir = "/home/ilya/Documents/IT-Geo-Paleochannel/data/river_dataset_test_1/masks_npy"


os.makedirs(npy_dir, exist_ok=True)

jpg_files = [f for f in os.listdir(jpg_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

for fname in jpg_files:
    jpg_path = os.path.join(jpg_dir, fname)
    npy_path = os.path.join(npy_dir, os.path.splitext(fname)[0] + '.npy')
    
    #    RGB,   512x512
    img = Image.open(jpg_path).convert('L').resize((512, 512), Image.BILINEAR)
    img_array = np.array(img)  # shape  (512, 512, 3)
    
    #   npy
    np.save(npy_path, img_array)
    print(f"[OK] {fname} -> {npy_path}, shape: {img_array.shape}")
```
